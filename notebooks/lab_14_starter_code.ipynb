## Bonus Section: Comparing Training and Test Performance

So far, we have evaluated how well our model performs on the test data.

We can also check how well the model performs on the training data. This helps us compare:

- How well the model performs on data it learned from
- How well the model performs on new, unseen data

In `sklearn`, regression models have a `.score()` method.

For linear regression, `.score()` returns the model's **R² score**.

R² measures how much of the variation in the target variable is explained by the model.

Run the following code:

```python
print(f"Training score: {model.score(X_train, y_train):.3f}")
print(f"Testing score: {model.score(X_test, y_test):.3f}")
```

### How to Interpret the Scores

If the training score is much higher than the testing score, the model may be overfitting.

This means the model performs well on the data it learned from, but does not perform as well on new, unseen data.

If the training score and testing score are both low, the model may be underfitting.

This means the model is not learning enough of the relationship between the features and the target variable.

If the training score and testing score are close together, and both scores are reasonably strong for the problem, the model may be generalizing well.

This means the model is learning useful patterns from the training data and still performing well on new data.
